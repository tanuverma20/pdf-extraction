```python
def extract_tables_from_pdf(pdf_path, output_folder):

    pdf_name = Path(f"/content/test3 (1).pdf").stem
    output_excel = os.path.join(output_folder, f"{pdf_name}.xlsx")

    with pdfplumber.open(f"/content/test3 (1).pdf") as pdf:
        writer = pd.ExcelWriter(output_excel, engine='openpyxl')
        sheet_count = 1
        tables_found = False

        for i, page in enumerate(pdf.pages):
            tables = page.extract_tables()

            if tables:
                tables_found = True
                for table in tables:
                    df = pd.DataFrame(table)
                    df.to_excel(writer, sheet_name=f"Sheet{sheet_count}", index=False, header=False)
                    sheet_count += 1


        if not tables_found:
            df = pd.DataFrame(["No tables found in this PDF."])
            df.to_excel(writer, sheet_name="No Tables", index=False, header=False)

        writer.close()

    print(f"✅ Extracted tables saved to: {output_excel}")
    return output_excel
```
